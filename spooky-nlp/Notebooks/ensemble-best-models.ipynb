```python
import pandas as pd
```

```python
combined_pred = pd.read_csv('../predictions/final_pred.csv')
short_pred = pd.read_csv('../predictions/short_pred.csv')
xgb2_pred = pd.read_csv('../predictions/xgboost2_pred.csv')
xgb3_pred = pd.read_csv('../predictions/xgb3_pred.csv')
```

```python
print(combined_pred.shape)
print(short_pred.shape)
print(xgb2_pred.shape)
print(xgb3_pred.shape)
```

```python
all_pred = pd.concat((combined_pred, short_pred, xgb2_pred, xgb3_pred))
```

```python
by_id_index = all_pred.groupby('id', as_index=False)
final_pred = by_id_index.mean()

```
